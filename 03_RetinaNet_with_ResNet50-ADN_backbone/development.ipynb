```python
import argparse

parser = argparse.ArgumentParser(description='PyTorch Detection Testing')
parser.add_argument('-f')
parser.add_argument('--model', type=str, default='retinanet_resnet50_fpn')
parser.add_argument('--weights-backbone', type=str, default='/home/hslee/INU_RISE/02_AdaptiveDepthNetwork/checkpoint/model_145.pth')
args = parser.parse_args()
```

```python
# check the backbone weights
backbone = torch.load(args.weights_backbone)

# get_state_dict()
```

```python
print(backbone['model'].keys())
```

```python
# check the number of parameters
num_params = 0
for k, v in backbone['model'].items():
    print(k, v.size())
    num_params += v.numel()
```

```python
# use the torch model reitnaet_resnet50_fpn with the custom backbone weights
from torchvision.models.detection import retinanet_resnet50_fpn

model = retinanet_resnet50_fpn(weights=None)
model.load_state_dict(backbone['model'], strict=False)
model.eval()
```
